```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, KFold
```

```python
ybin = pd.read_csv('./data/ybin.csv', header=None).values
```

```python
display(ybin)
```

```python
ninst, nalgos = ybin.shape
print('num instances', ninst)
print('num algos', nalgos)
```

```python
out = {
    'cp': [None] * nalgos,
}
```

```python
for i in range(nalgos):
    # skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=6)
    # out['cp'][i] = list(skf.split(np.zeros(ninst), ybin[:, i]))
    kf = KFold(n_splits=5, shuffle=True, random_state=10)
    out['cp'][i] = list(kf.split(np.zeros(ninst)))
    

    print(f'Ybin[:, {i}]:')
    print(ybin[:, i])
    
    # Print k-fold indices
    for jj, (train_idx, test_idx) in enumerate(out['cp'][i]):
        print(f'Algorithm {i+1}, Fold {jj+1}')
        print(f'Testing indices: {test_idx + 1}')
        print(f'Training indices: {train_idx + 1}')  
         
```
